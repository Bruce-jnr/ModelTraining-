```python
import json
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from transformers import AutoModelForSequenceClassification, AutoTokenizer # Import AutoTokenizer
from tqdm import tqdm
import torch # Import torch

# Load your dataset (update the path to your JSON file)
DATASET_PATH = "/content/drive/MyDrive/Colab Notebooks/vectara.json"

with open(DATASET_PATH, "r") as f:
    data = json.load(f)

# Prepare input pairs and true labels
pairs = [(item["source"], item["claim"]) for item in data]
true_labels = [item["label"] for item in data]  # 0 = hallucinated, 1 = factual

# Load the Vectara model and tokenizer
print("Loading Vectara model and tokenizer...")
model = AutoModelForSequenceClassification.from_pretrained(
    'vectara/hallucination_evaluation_model', trust_remote_code=True
)
tokenizer = AutoTokenizer.from_pretrained(
    'vectara/hallucination_evaluation_model', trust_remote_code=True
)


# Get model scores
print("Scoring data using Vectara...")
scores = []
batch_size = 32

for i in tqdm(range(0, len(pairs), batch_size)):
    batch = pairs[i:i+batch_size]
    # Prepare input for the model
    inputs = tokenizer(batch, padding=True, truncation=True, return_tensors="pt")
    with torch.no_grad():
        outputs = model(**inputs)
    # The model outputs logits, apply sigmoid to get probabilities
    batch_scores = torch.sigmoid(outputs.logits).squeeze().tolist()
    scores.extend(batch_scores)

# Convert probability scores to predicted labels
threshold = 0.5
predicted_labels = [1 if score >= threshold else 0 for score in scores]

# Evaluate
accuracy = accuracy_score(true_labels, predicted_labels)
precision = precision_score(true_labels, predicted_labels)
recall = recall_score(true_labels, predicted_labels)
f1 = f1_score(true_labels, predicted_labels)

print("\nðŸ“Š Evaluation Results on TruthfulQA:")
print(f"Accuracy:  {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall:    {recall:.4f}")
print(f"F1 Score:  {f1:.4f}")
```

```python
from transformers import pipeline, AutoTokenizer
import json
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from tqdm import tqdm
import torch

# Load your dataset (update the path to your JSON file)
DATASET_PATH = "/content/drive/MyDrive/Colab Notebooks/vectara.json"

with open(DATASET_PATH, "r") as f:
    data = json.load(f)

# Prepare input pairs and true labels from your dataset
pairs = [(item["source"], item["claim"]) for item in data]

# Map string labels to integers during data loading
label_mapping = {'hallucinated': 0, 'factual': 1, 'consistent': 1} # Assuming 'consistent' also maps to 1
true_labels = [label_mapping.get(str(item["label"]).lower(), -1) for item in data] # Convert label to string and lowercase for robust mapping, use -1 for unknown labels

# Remove entries with unknown labels if any
valid_indices = [i for i, label in enumerate(true_labels) if label != -1]
pairs = [pairs[i] for i in valid_indices]
true_labels = [true_labels[i] for i in valid_indices]


# Prompt the pairs
prompt = "<pad> Determine if the hypothesis is true given the premise?\n\nPremise: {text1}\n\nHypothesis: {text2}"
input_pairs = [prompt.format(text1=pair[0], text2=pair[1]) for pair in pairs]

# Use text-classification pipeline to predict
print("Loading pipeline and scoring data...")
classifier = pipeline(
            "text-classification",
            model='vectara/hallucination_evaluation_model',
            tokenizer=AutoTokenizer.from_pretrained('google/flan-t5-base'),
            trust_remote_code=True
        )
full_scores = classifier(input_pairs, top_k=None) # List[List[Dict[str, float]]]

# Extract the scores for the 'consistent' label (assuming 'consistent' corresponds to factual)
# Note: The model's output labels might need to be confirmed from its documentation.
# Assuming 'consistent' label score indicates the probability of being factual (label 1)
scores = [score_dict['score'] for score_for_both_labels in full_scores for score_dict in score_for_both_labels if score_dict['label'] == 'consistent']


# Convert probability scores to predicted labels
# If 'consistent' score is high, predict label 1 (factual), otherwise 0 (hallucinated)
threshold = 0.5
predicted_labels = [1 if score >= threshold else 0 for score in scores]

# Ensure labels are integers for scikit-learn metrics (already done during mapping, but keeping for clarity)
true_labels = [int(label) for label in true_labels]
predicted_labels = [int(label) for label in predicted_labels]


# Evaluate the model's performance
print("\nðŸ“Š Evaluation Results:")
accuracy = accuracy_score(true_labels, predicted_labels)
precision = precision_score(true_labels, predicted_labels)
recall = recall_score(true_labels, predicted_labels)
f1 = f1_score(true_labels, predicted_labels)

print(f"Accuracy:  {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall:    {recall:.4f}")
print(f"F1 Score:  {f1:.4f}")
```
